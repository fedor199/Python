#### Задание 3, задача 1
Напишите код, который будет переводить полярные координаты в декартовы.

```python
r = float(input('Введите радиус: '))
alpha = np.pi/3
x = r * np.cos(alpha)
y = r * np.sin(alpha)
plt.scatter(x, y, color='red')
print("x = {:.2f}, y = {:.2f}".format(x, y))
plt.xlim(0, 5)
plt.ylim(0, 5)
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
```

#### Задание 3, задача 2
Напишите код, который будет рисовать график окружности в полярных координатах.

```python
x = []
y = []
R = 6.0
k = -R
while k <= R:
    x.append(k)
    y.append(R)
    k += 0.0001
plt.polar(x, y, color='blue')
plt.show()
```

#### Задание 4, пример 1
Решите систему уравнений:

y = x2 – 1

exp(x) + x∙(1 – y) = 1

```python
from scipy.optimize import fsolve

def equations(p):
    x, y = p
    return (y - x**2 + 1, (y - (np.exp(x) + x - 1)/x))

x = np.linspace(-2, 5,201)
plt.plot(x, (np.exp(x) + x - 1)/x)
plt.plot(x, x**2 - 1)
x1, y1 = fsolve(equations, (-2, 1))
x2, y2 = fsolve(equations, (3, 5))
x3, y3 = fsolve(equations, (4, 5))
print(x1, y1)
print(x2, y2)
print(x3,y3)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-2,20)
plt.grid(True)
plt.plot(x1, y1, 'o', color='green')
plt.plot(x2, y2, 'o', color='green')
plt.plot(x3, y3, 'o', color='green')
plt.show()
```

#### Задание 4, пример 2
Решите систему уравнений и неравенств:

y = x2 – 1

exp(x) + x∙(1 – y) > 1


#### Область решения системы выделена зеленым цветом на графике

```python
x = []
y = []
x_= []
y_= []
i = -7 
while i <= 7:
    f = np.exp(i) + i * (2 - i**2)
    if f > 1:
        x.append(i)
        y.append(f)
    else:
        x_.append(i)
        y_.append(f)
    i+=0.01
plt.grid(True)
plt.ylim(-5,10)
plt.scatter(x, y, color='green')
plt.scatter(x_, y_, color='red')
plt.show()
```
